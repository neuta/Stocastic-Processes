```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
norm.pdf(x=1.0, loc=0, scale=1)
```

```python
fig, ax = plt.subplots()
x = np.linspace(-6,6,100)
ax.plot(x, norm.pdf(x, loc=0, scale=1), linewidth=2, color='m')
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
plt.savefig('funcionnormal', format='jpeg', dpi=200)
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
stdvs = [1.0, 2.0, 3.0, 4.0]
for s in stdvs:
    ax.plot(x, norm.pdf(x,loc=1,scale=s), label='stdv=%.1f' % s)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
```

```python
fig, ax = plt.subplots()
x = np.linspace(-10,10,100)
means = [0.0, 1.0, 2.0, 5.0]
for mean in means:
    ax.plot(x, norm.pdf(x,loc=mean,scale=1), label='mean=%.1f' % mean)
    
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.set_title('Normal Distribution')
ax.legend(loc='best', frameon=True)
ax.set_ylim(0,0.45)
ax.grid(True)
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-6,6,0.001)
ax.plot(x, norm.pdf(x),'m')
ax.set_title("Cumulative normal distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for filling
px=np.arange(-6,1,0.01)
ax.set_ylim(0,0.5)
ax.fill_between(px,norm.pdf(px),alpha=0.5, color='m')
plt.savefig('funcioacumulada', format='jpeg', dpi=200)
plt.show()
```

```python
# Given the mean of 3 and the standard deviation of 2, we can find the probability of 𝑃(𝑋<2).
from scipy.stats import norm
mean = 5
std  = 2
xto  = 3.5
lessthan2 = norm.cdf(x=xto, loc=mean, scale=std)
print(lessthan2)
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-6,15,0.001)
ax.plot(x, norm.pdf(x,loc=mean,scale=std),'m')
ax.set_title("N(5,$2^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for fill_between
px=np.arange(-6,xto,0.01)
ax.set_ylim(0,0.25)
ax.fill_between(px,norm.pdf(px,loc=mean,scale=std),alpha=0.5, color='m')
# for text
ax.text(1,0.01,round(lessthan2,2), fontsize=12)
plt.savefig('funcioacumuladaejemplo', format='jpeg', dpi=300)
plt.show()
```

```python
from scipy.stats import norm
mean = 0
std  = 1
xto  = 3*std
lessthan2 = norm.cdf(x=xto, loc=mean, scale=std)
print(lessthan2)
print(1-lessthan2)
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-6,6,0.001)
ax.plot(x, norm.pdf(x,loc=mean,scale=std),'m')
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for fill_between
px=np.arange(xto,4,0.01)
ax.set_ylim(0,0.42)
ax.fill_between(px,norm.pdf(px,loc=mean,scale=std),alpha=0.5, color='g')
# for text
plt.savefig('funcioacumuladaejemplo2', format='jpeg', dpi=300)
plt.show()
```

```python
#Caso de evaluación de probabilidad mayor a 1sigmas, 2sigmas, 3sigmas, 4sigmas
from scipy.stats import norm
k = [1.0, 2.0, 3.0, 3.1, 3.2, 3.3, 3.4, 3.5]
mean = 0
std  = 1

lessthan2 = norm.cdf(x=k, loc=mean, scale=std)
sk = 1-lessthan2
print(sk)
```

```python
#la probabilidad de dos eventos es igual al producto, en este caso al producto de probabilidades de 1sigmas, 2sigmas, 3sigmas, 4sigmas y 5sigmas
sk2 = sk*sk
sk2
```

```python
#Caso de evaluación de probabilidad mayor a 2sigmas, 4sigmas, 6sigmas, 8sigmas
from scipy.stats import norm
k2 = [2.0, 4.0, 6.0, 6.2, 6.4, 6.6, 6.8, 7.0]
mean = 0
std  = 1

lessthan2 = norm.cdf(x=k2, loc=mean, scale=std)
s2k = 1-lessthan2
print(s2k)
```

```python
#Para comparar Ratio of two occurrences of size K by one of 2K for a Gaussian distribution
Ratio = sk2/s2k
Ratio
```

```python
fig, ax = plt.subplots()
x= np.arange(0,4,0.001)
ax.plot(k, Ratio,'m')
ax.set_ylim(0,45000)
ax.grid(True)
ax.set_xlabel('K(std)')
ax.set_ylabel('S(K) / S(2K)')
plt.savefig('RatioS2k', format='jpeg', dpi=300)
plt.show()
```
